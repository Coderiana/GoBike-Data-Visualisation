## When the bikes are used?

The most busy time of the day is in the morning between 7 and 9 am and by the evening between 5 and 7 pm. This points out towards the cycle-to-work sheme and it is evident that majority of users is taking the bike to commute to and from work. This is a great find that GoBike is used to help professionals commute to work. This user group seems to be the root of the service, not the leisure bikers.
